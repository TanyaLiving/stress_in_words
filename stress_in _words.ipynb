```python
def S_A(series):
    if series['pos'] in (['S', 'A']):
        if re.search('\(', series['gr']) is not None:
            return re.search('\(.*\)', series['gr']).group().strip('\(\)').split('|')
        else:
            return re.search('=(.*)', series['gr']).group(1)
    else:
        return 'not S_A'

def V(series):
    if series['pos'] == 'V':
        text = re.search('=(.*)', series['gr']).group(1).split(',')
        return list(set(verb_tense) & set(text))
    else:
        return 'not_a_verb'
```
